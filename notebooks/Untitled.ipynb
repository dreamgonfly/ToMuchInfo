```python
from gensim.models import FastText
from konlpy.tag import Twitter
```

```python
import pandas as pd
```

```python
with open("../code/data/small/train/train_data") as f:
    data = f.readlines()
    data = [str(t.strip()) for t in data]
    data = pd.DataFrame(data, columns=['comment'])
    
with open("../code/data/small/train/train_label") as f:
    score = f.readlines()
    data['score'] = pd.DataFrame(score)
    data['score'] = data['score'].map(lambda x: int(x.strip()))
```

```python
tokenizer = Twitter()
```

```python
reviews = data['comment'].values
```

```python
tokenized_reviews = [tokenizer.pos(review, stem=False) for review in reviews]
```

```python
tokens = [[token for token, pos in tokenized_list] for tokenized_list in tokenized_reviews]
tags = [[pos for token, pos in tokenized_list] for tokenized_list in tokenized_reviews]
```

```python
fasttext = FastText(sentences=tokens,
                         sg=1,
                         window=5,
                         negative=10,
                         min_n=2,
                         max_n=5,
                         word_ngrams=1,
                         size=300,
                         iter=20,
                         sorted_vocab=1,
                         )
```

```python
vocab_words = list(fasttext.wv.vocab.keys())
word2idx = {word: idx for idx, word in enumerate(vocab_words)}
word2idx['<UNK>'] = len(vocab_words)
word2idx['<PAD>'] = len(vocab_words) + 1

idx2word = {idx: word for idx, word in enumerate(vocab_words)}
idx2word[len(vocab_words)] = '<UNK>'
idx2word[len(vocab_words) + 1] = '<PAD>'
print("word2idx", word2idx['쓰레기'], word2idx['<PAD>'])
print("idx2word", idx2word[word2idx['쓰레기']], idx2word[word2idx['<PAD>']])
print("vocab_words", vocab_words[0])
```

```python
word_vectors = []
vocab_num = len(vocab_words)
for i in range(20000):
    if i < vocab_num:
        word = idx2word[i]
        if word in ['<UNK>', '<PAD>']:
            vector = np.zeros(300)
        else:
            vector = fasttext.wv[word]
        word_vectors.append(vector)
    else:
        word_vectors.append(np.zeros(300))
embedding = np.stack(word_vectors)
```
